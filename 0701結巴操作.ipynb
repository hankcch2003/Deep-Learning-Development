```python
import jieba
text1='我要成為神奇寶貝大師'
text2='我家門前有條水溝很難過'
text3='電視主播傅達仁罹患末期癌症'
print('分詞部分')
jieba1 = jieba.cut(text1)
print('/'.join(jieba1))
jieba2 = jieba.cut(text2)
print('/'.join(jieba2))
jieba3 = jieba.cut(text3)
print('/'.join(jieba3))
```

```python
import jieba
text4='蔡英文總統今（8）日上午前往臺中出席「臺灣國立大學系統啟動典禮」'
text5='賴清德總統今（6）日上午前往桃園出席「中央警察大學114年聯合畢業典禮」'
print('分詞部分')
jieba4 = jieba.cut(text4)
print('/'.join(jieba4))
jieba5 = jieba.cut(text5)
print('/'.join(jieba5))
import jieba.posseg as pseg
print('詞性部分')
jieba4 = pseg.cut(text4)
print('第4個')
for word, flag in jieba4:
    print(word, flag)
jieba5 = pseg.cut(text5)
print('第5個')
for word, flag in jieba5:
    print(word, flag)
print('想要抓取實體名稱，jieba的詞性是nr ns nt 都是實體')
print('詞性部分')
jieba4 = pseg.cut(text4)
print('第4個')
for word, flag in jieba4:
    if flag in ['nr','ns','nt']:
        print(word, flag)
jieba5 = pseg.cut(text5)
print('第5個')
for word, flag in jieba5:
    if flag in ['nr','ns','nt']:    
        print(word, flag)
```

```python
import jieba
print('====如果有些詞彙沒列入語助詞stopWords.txt，請於txt檔案內新增====')
list1 = []
stopWords = []  # 不要的詞彙
remainderWords = []  # 去除不要的詞彙的清單
with open('stopWords.txt','r', encoding='utf-8') as file:
    for data in file:
        data = data.strip()  # 去除空白
        stopWords.append(data)
with open('data.txt','r', encoding='utf-8') as file:
    text = file.read()
    segment = jieba.cut(text)
print('不要的清單有了，我們的文字清單也有了，接著就要進行過濾')
remainderWords = list(filter(lambda a:a not in stopWords and a!='\n', segment))
print('總共有多少詞彙', len(remainderWords))
print('詞彙統計')
import collections
counter1 = collections.Counter(remainderWords)
print('找前10個詞彙')
for key, values in counter1.most_common(10):
    print(key,":", values)
```

```python
print(remainderWords)
from wordcloud import WordCloud
import matplotlib.pyplot as plt
text1 = ' '.join(remainderWords)
wc = WordCloud(font_path='NotoSerifCJKtc-Black.otf', width = 800, height = 600, background_color='white').generate(text1)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
