```python
print('前面是訓練資料，後續是測試資料')
print('0-如果是csv檔案，會先載入資料、檢視資料然後去除遺失值')
mlptest1=pd.read_csv('/content/drive/MyDrive/Colab第三階段202505/ch2_20250603/MLPtest1.csv')
print(mlptest1.shape)
print(mlptest1.head())
del mlptest1['Unnamed: 0']
print(mlptest1.head())
print('檢視資料與資料型態')
mlptest1.to_html('mlptest1.html') # 轉換為網頁
print(mlptest1.dtypes) # 檢視欄位資料型態
print('檢視欄位型態為float或int，若是object得做編碼轉換')
print('查看是否有遺失值')
print(mlptest1.isnull().sum())
print('MLPtest2.csv主要是要說明訓練資料，也就是MLPtest1的答案y_train的來源')
print('所以MLPtest2.csv於此是做參考')
mlptest2=pd.read_csv('/content/drive/MyDrive/Colab第三階段202505/ch2_20250603/MLPtest2.csv')
col_names=['y1','y2','y3','y4','y5','y6','y7','y8','y9','y10']
print(mlptest2.shape)
print(mlptest2.head())
del mlptest2['Unnamed: 0']
mlptest2.columns=col_names
print(mlptest2.head())
```

```python
print('1-切割訓練與測試資料')
x_train=mlp1.iloc[:,:-1]
y_train=mlp1.iloc[:,-1]
print(x_train.shape)
print(y_train.shape)
x_test=mlptest1.iloc[:,:-1]
y_test=mlptest1.iloc[:,-1]
print(x_test.shape)
print(y_test.shape)
import tensorflow as tf
category=10 # 分類目標內容數量
dim=2 # 特徵欄位數量
print('2-模型規劃')
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=10,activation='relu',input_shape=(dim,))) # 輸入
model.add(tf.keras.layers.Dense(units=category,activation='relu')) # 隱藏層
model.add(tf.keras.layers.Dense(units=category,activation='softmax')) # 最終結果
print(model.summary())
print('3-模型編譯')
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
print('4-訓練模型')
model.fit(x_train,y_train,epochs=100,batch_size=128)
print('5-評估模型')
loss,acc=model.evaluate(x_test,y_test)
print('loss=',loss)
print('acc=',acc)
print('6-預測結果')
y_pred=model.predict(x_test)
print(y_pred)
```
