```python
import jieba
print('====就職演說中有哪些詞彙，共用了幾次?====')
list1 = []
with open('data.txt','r', encoding='utf-8') as file:
    text = file.read()
    # print(text)
    segment = jieba.cut(text)
    print('====分詞====')
    for seg in segment:
        list1.append(seg)
        # print(seg)
print('總共有多少詞彙', len(list1))
print('詞彙統計')
import collections
counter1 = collections.Counter(list1)
print('找前20個詞彙')
for key, values in counter1.most_common(20):
    print(key,":", values)
```
