```python
import os, time
import jieba
t1 = time.time()

def read_files(file_path):
  texts = []
  for file_name in os.listdir(file_path):
    file_path_name = os.path.join(file_path, file_name)
    if file_name.endswith('.txt'):
      with open(file_path_name, 'r', encoding='utf-8') as f:
        texts.append(f.read())
  return texts

def preprocess_text(text):
  return ' '.join(jieba.cut(text))
file_path = '/content/drive/MyDrive/Colab第三階段202505/ch13_20250710/CM'
texts = read_files(file_path)
print(texts)
processed_texts = [preprocess_text(text) for text in texts]
print('分詞後:', processed_texts)
print('TfidfVectorizer')

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(processed_texts)
print(X.toarray())
words = vectorizer.get_feature_names_out()
print(words)

import pandas as pd
tfidf_df = pd.DataFrame(X.toarray(), columns = words)
print(tfidf_df)
print('tfidf_df 形狀:\n', tfidf_df.shape)
tfidf_df.to_html('tfidf_df.html')

def get_top_kewords(row, words, top_n = 10):  # 接收三個，前一個接收兩個
  top_kewords = row.sort_values(ascending = False).head(top_n)
  return list(zip(top_kewords.index, top_kewords.values))

doc_id = 0
top_kewords = get_top_kewords(tfidf_df.iloc[doc_id], words)
print('文件ID:', doc_id, '中TFIDF權重最高的詞彙:')

for ngram, count in top_kewords:
  print(ngram, ':', count)
t2 = time.time()
print('執行時間:', t2 - t1)
```
